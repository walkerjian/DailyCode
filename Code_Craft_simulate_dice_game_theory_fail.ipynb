##Solution:
To help Alice decide which game to elect to play based on their expected values, we can simulate both games using Python. The expected value is the average amount she will pay if she plays the game a very large number of times.

Here's a Python program to simulate both games. We will simulate each game 100,000 times to get a good approximation of their expected values.

1. **Game 1**: The player rolls a die until they get a five followed by a six.
2. **Game 2**: The player rolls a die until they get a five followed by another five.


##Implementation:
Let's simulate these games and calculate their expected values:

```python
import random

def simulate_game(target_sequence):
    count = 0
    total_rolls = 0
    num_simulations = 100000
    for _ in range(num_simulations):
        sequence = []
        while True:
            roll = random.randint(1, 6)
            total_rolls += 1
            sequence.append(roll)
            # Check if the last two rolls match the target sequence
            if sequence[-len(target_sequence):] == target_sequence:
                break
        count += len(sequence)
    return total_rolls / num_simulations

# Game 1: Target sequence is [5, 6]
expected_value_game1 = simulate_game([5, 6])

# Game 2: Target sequence is [5, 5]
expected_value_game2 = simulate_game([5, 5])

expected_value_game1, expected_value_game2
```
