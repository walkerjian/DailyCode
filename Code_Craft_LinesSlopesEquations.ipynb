To solve for the lengths marked $ a $ and $ b $, we'll use the properties of similar triangles. The three triangles shown are right-angled and similar to each other, which means their corresponding sides are proportional.

From the smallest triangle, we have the sides:
- Vertical side: 6 units
- Horizontal side: 8 units

From the middle triangle, we have the vertical side as 9 units, and we need to find the horizontal side, which is $ a $.

From the largest triangle, we want to find the vertical side, $ b $, and we know the horizontal side is 10 units.

Since the triangles are similar, the ratios of the corresponding sides will be equal. That is:
$
\frac{\text{Vertical side of the smallest triangle}}{\text{Horizontal side of the smallest triangle}} = \frac{\text{Vertical side of the middle triangle}}{\text{Horizontal side of the middle triangle}} = \frac{\text{Vertical side of the largest triangle}}{\text{Horizontal side of the largest triangle}}
$

We can set up the proportions as follows:
$
\frac{6}{8} = \frac{9}{a} \quad \text{and} \quad \frac{6}{8} = \frac{b}{10}
$

We can solve these equations to find $ a $ and $ b $.

The lengths marked $ a $ and $ b $ in the diagram are $ 12 $ units and $ 7.5 $ units, respectively.
