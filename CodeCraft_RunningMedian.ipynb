#Problem:
Compute the running median of a sequence of numbers. That is, given a stream of numbers, print out the median of the list so far on each new element.

Recall that the median of an even-numbered list is the average of the two middle numbers.

For example, given the sequence [2, 1, 5, 7, 2, 0, 5], your algorithm should print out:

2
1.5
2
3.5
2
2
2

#Solution:
To compute the running median of a sequence of numbers efficiently, we can use two heaps (or priority queues):

1. A max heap to store the smaller half of the numbers.
2. A min heap to store the larger half of the numbers.

The max heap will store the smaller half of the numbers in reverse order (largest at the root), and the min heap will store the larger half of the numbers in ascending order (smallest at the root).

### Steps:
- Insert the new number into one of the heaps.
- Balance the heaps such that the max heap either has the same number of elements as the min heap or one more element than the min heap.
- The median will be either:
  - The root of the max heap if it has more elements, or
  - The average of the roots of the max heap and the min heap if they have the same number of elements.

Here's how we can implement this:

```python
import heapq

class RunningMedian:
    def __init__(self):
        # Max heap for the smaller half, stored as negative values for heapq to act as max heap
        self.small = []
        # Min heap for the larger half
        self.large = []

    def add_number(self, num):
        # Add to max heap (small), but store as negative to simulate a max heap
        heapq.heappush(self.small, -num)
        
        # Ensure the max heap has smaller elements and the min heap has larger elements
        if self.small and self.large and (-self.small[0] > self.large[0]):
            heapq.heappush(self.large, -heapq.heappop(self.small))

        # Balance the heaps: max heap can have at most one extra element
        if len(self.small) > len(self.large) + 1:
            heapq.heappush(self.large, -heapq.heappop(self.small))
        elif len(self.large) > len(self.small):
            heapq.heappush(self.small, -heapq.heappop(self.large))

    def get_median(self):
        # If max heap has more elements, the median is the root of max heap
        if len(self.small) > len(self.large):
            return -self.small[0]
        # If both heaps have the same number of elements, the median is the average of the two roots
        else:
            return (-self.small[0] + self.large[0]) / 2

def print_running_median(sequence):
    rm = RunningMedian()
    for num in sequence:
        rm.add_number(num)
        print(rm.get_median())

# Test with the given sequence
sequence = [2, 1, 5, 7, 2, 0, 5]
print_running_median(sequence)
```

### Explanation:
- We use two heaps, `self.small` (max heap) and `self.large` (min heap).
- Each time a new number is added, it is first pushed to the max heap (`self.small`), and then we balance the heaps to maintain the properties of the heaps (max heap for smaller half and min heap for larger half).
- After each insertion, we compute the median:
  - If `self.small` has more elements, the median is the root of `self.small`.
  - If both heaps are of the same size, the median is the average of the two roots.

### Output for the sequence `[2, 1, 5, 7, 2, 0, 5]`:

```

2

2
2
2
```

````python
import heapq

# Model: Contains the core logic for computing the running median
class RunningMedianModel:
    def __init__(self):
        # Max heap for the smaller half of numbers
        self.small = []
        # Min heap for the larger half of numbers
        self.large = []

    def add_number(self, num):
        try:
            # Add to max heap (small), but store as negative to simulate a max heap
            heapq.heappush(self.small, -num)

            # Ensure the max heap has smaller elements and the min heap has larger elements
            if self.small and self.large and (-self.small[0] > self.large[0]):
                heapq.heappush(self.large, -heapq.heappop(self.small))

            # Balance the heaps: max heap can have at most one extra element
            if len(self.small) > len(self.large) + 1:
                heapq.heappush(self.large, -heapq.heappop(self.small))
            elif len(self.large) > len(self.small):
                heapq.heappush(self.small, -heapq.heappop(self.large))

        except TypeError:
            raise ValueError("Invalid input: only numbers are allowed")

    def get_median(self):
        if not self.small and not self.large:
            raise ValueError("No elements in the sequence")

        if len(self.small) > len(self.large):
            return -self.small[0]
        else:
            return (-self.small[0] + self.large[0]) / 2

# View: Responsible for displaying results and error messages
class RunningMedianView:
    @staticmethod
    def display_median(median):
        print(f"Current Median: {median}")

    @staticmethod
    def display_error(error_message):
        print(f"Error: {error_message}")

# Controller: Manages the flow between Model and View
class RunningMedianController:
    def __init__(self):
        self.model = RunningMedianModel()
        self.view = RunningMedianView()

    def process_sequence(self, sequence):
        for num in sequence:
            try:
                self.model.add_number(num)
                median = self.model.get_median()
                self.view.display_median(median)
            except ValueError as e:
                self.view.display_error(str(e))
                return  # Stop further processing on invalid input

# Test Cases
def run_tests():
    controller = RunningMedianController()

    # Test 1: Normal sequence
    print("Test 1: Normal sequence")
    try:
        controller.process_sequence([2, 1, 5, 7, 2, 0, 5])
    except Exception as e:
        print(f"Test 1 failed: {e}")

    # Test 2: Single element
    print("\nTest 2: Single element")
    try:
        controller.process_sequence([42])
    except Exception as e:
        print(f"Test 2 failed: {e}")

    # Test 3: Empty input
    print("\nTest 3: Empty input")
    try:
        controller.process_sequence([])
    except Exception as e:
        print(f"Test 3 failed: {e}")

    # Test 4: Non-integer input (should raise error)
    print("\nTest 4: Non-integer input")
    try:
        controller.process_sequence([2, 1, "a", 7])
    except Exception as e:
        print(f"Test 4 failed: {e}")

    # Test 5: Large dataset (performance test)
    print("\nTest 5: Large dataset")
    try:
        large_sequence = list(range(1000))  # Sequence from 0 to 999
        controller.process_sequence(large_sequence)
    except Exception as e:
        print(f"Test 5 failed: {e}")

# Execute all tests
run_tests()
````
