# Reversal of a Directed Graph (Digraph)

## Problem:

An algorithm that computes the reversal of a directed graph. Given a directed graph, the reversal of the graph is a new graph that has the same nodes and edges, but the direction of every edge is reversed.

## Definition:

### Directed Graph:

A directed graph (or digraph) is a set of vertices and a collection of directed edges (also called arcs) that each connects an ordered pair of vertices. The order of the pair matters, meaning \(A to B\) is different from \(B to A\).

### Use of Directed Graphs:

Directed graphs are used in a variety of fields and applications, including:
- Representing finite state machines.
- Modeling processes in which information or objects flow in a consistent direction (e.g., computer networks, traffic flow).
- Dependency graphs where one task depends on the completion of another.
- Representing hierarchies and taxonomies.

## Solution:

1. **Model**: The model will represent our directed graph and its reversed version.
2. **View**: The view will be responsible for showing the graphical representation of our graphs.
3. **Controller**: The controller will handle the logic for reversing the graph and managing the interaction between the model and the view.

Here's a breakdown of the solution:

### Model (`GraphModel`):
- Represents the data structure of our directed graph.
- Provides methods to add edges and reverse the graph.

### View (`GraphView`):
- Represents the graphical representation of the graph.
- Uses the NetworkX and Matplotlib libraries to visualize the graph.

### Controller (`GraphController`):
- Manages the interaction between the model and the view.
- Provides an interface to add edges to the graph.
- Uses the view to display the original and reversed graphs.

### Test Function (`test_reversal`):
- Contains 10 test cases to validate the graph reversal algorithm.
- For each test case, it initializes the MVC components and adds the edges to the graph.
- It then uses the controller to display both the original and reversed graphs.


```python
import networkx as nx
import matplotlib.pyplot as plt


class GraphModel:
    """Represents the directed graph model."""

    def __init__(self):
        """Initializes an empty directed graph."""
        self.graph = nx.DiGraph()

    def add_edge(self, start, end):
        """Add a directed edge to the graph."""
        self.graph.add_edge(start, end)

    def reverse(self):
        """Returns the reversed graph."""
        return self.graph.reverse()


class GraphView:
    """Represents the graphical representation of the graph."""

    @staticmethod
    def plot_graph(graph, title):
        """Plots the graph with a given title."""
        pos = nx.spring_layout(graph)
        nx.draw(graph, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=20)
        plt.title(title)
        plt.show()


class GraphController:
    """Handles the logic for reversing the graph and interaction between model and view."""

    def __init__(self, model, view):
        """Initialize the controller with a model and view."""
        self.model = model
        self.view = view

    def add_edge(self, start, end):
        """Add a directed edge to the graph."""
        self.model.add_edge(start, end)

    def display_graph(self):
        """Display the original and reversed graph."""
        self.view.plot_graph(self.model.graph, "Original Graph")
        reversed_graph = self.model.reverse()
        self.view.plot_graph(reversed_graph, "Reversed Graph")


def test_reversal():
    """Test function to validate the directed graph reversal algorithm."""

    test_cases = [
        [("A", "B"), ("B", "C")],
        [("A", "B"), ("B", "C"), ("C", "D")],
        [("A", "D"), ("D", "C"), ("C", "B"), ("B", "A")],
        [("M", "N"), ("N", "O")],
        [("X", "Y")],
        [("J", "K"), ("K", "L"), ("L", "M"), ("M", "N")],
        [("Z", "Y"), ("Y", "X"), ("X", "W")],
        [("G", "H"), ("H", "I"), ("I", "J")],
        [("P", "Q"), ("Q", "R"), ("R", "S")],
        [("U", "V"), ("V", "W"), ("W", "X"), ("X", "Y"), ("Y", "Z")]
    ]

    for edges in test_cases:
        model = GraphModel()
        view = GraphView()
        controller = GraphController(model, view)

        for edge in edges:
            controller.add_edge(edge[0], edge[1])

        controller.display_graph()


# Running the test function to validate the solution
test_reversal()
```

# Digraph Algebra

Algebra on directed graphs (or digraphs) is a rich area of study that touches various mathematical disciplines, including graph theory, linear algebra, and group theory.

## Algebra on Directed Graphs:

1. **Union:**
    - Given two digraphs \( G_1 \) and \( G_2 \), their union \( G_1 \cup G_2 \) is the graph with vertex set \( V(G_1) \cup V(G_2) \) and edge set \( E(G_1) \cup E(G_2) \).
   
2. **Intersection:**
    - Given two digraphs \( G_1 \) and \( G_2 \), their intersection \( G_1 \cap G_2 \) is the graph with vertex set \( V(G_1) \cap V(G_2) \) and edge set \( E(G_1) \cap E(G_2) \).

3. **Difference:**
    - Given two digraphs \( G_1 \) and \( G_2 \), the difference \( G_1 - G_2 \) is the graph obtained by removing from \( G_1 \) all the edges that are also in \( G_2 \) and all the isolated vertices.

4. **Complement:**
    - The complement of a digraph \( G \) is a graph on the same vertex set but with edge set consisting of the edges not in \( G \). So if there's an edge from \( A \) to \( B \) in \( G \), there isn't one in the complement, and vice versa.

5. **Cartesian Product:**
    - Given two digraphs \( G_1 \) and \( G_2 \), their cartesian product \( G_1 \times G_2 \) is the graph whose vertex set is the cartesian product of their vertex sets (pairs \( (v_1, v_2) \) where \( v_1 \in V(G_1) \) and \( v_2 \in V(G_2) \)) and there's an edge from \( (a_1, a_2) \) to \( (b_1, b_2) \) if and only if \( a_1 = b_1 \) and \( (a_2, b_2) \) is an edge in \( G_2 \), or \( (a_1, b_1) \) is an edge in \( G_1 \) and \( a_2 = b_2 \).

6. **Tensor Product (or Kronecker Product):**
    - Given two digraphs \( G_1 \) and \( G_2 \), their tensor product \( G_1 \otimes G_2 \) is the graph whose vertex set is the cartesian product of their vertex sets, and there's an edge from \( (a_1, a_2) \) to \( (b_1, b_2) \) if and only if \( (a_1, b_1) \) is an edge in \( G_1 \) and \( (a_2, b_2) \) is an edge in \( G_2 \).

7. **Homomorphisms and Isomorphisms:**
    - A homomorphism from a digraph \( G_1 \) to a digraph \( G_2 \) is a function from the vertex set of \( G_1 \) to the vertex set of \( G_2 \) that preserves edges. If this function is bijective and its inverse is also a homomorphism, then it's an isomorphism, and the graphs are said to be isomorphic.

8. **Graph Powers:**
    - The \( k \)-th power \( G^k \) of a digraph \( G \) has the same vertex set as \( G \), and there's an edge from \( a \) to \( b \) if and only if the shortest path from \( a \) to \( b \) in \( G \) has at most \( k \) edges.

### Implementing the Algebra:

Implementing the above operations programmatically is feasible and can be quite efficient for many practical purposes, especially if we leverage existing libraries like NetworkX. Note, however, that some operations, like finding isomorphisms between two graphs, can be computationally hard (NP-complete).


```python
class GraphAlgebra:
    """Defines and implements algebraic operations for directed graphs using NetworkX."""

    @staticmethod
    def union(G1, G2):
        """Returns the union of two directed graphs."""
        return nx.compose(G1, G2)

    @staticmethod
    def intersection(G1, G2):
        """Returns the intersection of two directed graphs."""
        intersected_graph = nx.DiGraph()
        for node in set(G1.nodes()).intersection(G2.nodes()):
            intersected_graph.add_node(node)
        for edge in set(G1.edges()).intersection(G2.edges()):
            intersected_graph.add_edge(*edge)
        return intersected_graph

    @staticmethod
    def difference(G1, G2):
        """Returns the difference between two directed graphs."""
        difference_graph = G1.copy()
        for edge in G2.edges():
            if edge in difference_graph.edges():
                difference_graph.remove_edge(*edge)
        for node in list(difference_graph.nodes()):
            if difference_graph.degree(node) == 0:
                difference_graph.remove_node(node)
        return difference_graph

    @staticmethod
    def complement(G1):
        """Returns the complement of a directed graph."""
        complement_graph = nx.DiGraph()
        nodes = list(G1.nodes())
        complement_graph.add_nodes_from(nodes)
        for i in range(len(nodes)):
            for j in range(len(nodes)):
                if i != j and not G1.has_edge(nodes[i], nodes[j]):
                    complement_graph.add_edge(nodes[i], nodes[j])
        return complement_graph

    @staticmethod
    def cartesian_product(G1, G2):
        """Returns the cartesian product of two directed graphs."""
        return nx.cartesian_product(G1, G2)

    @staticmethod
    def tensor_product(G1, G2):
        """Returns the tensor (Kronecker) product of two directed graphs."""
        return nx.tensor_product(G1, G2)

    @staticmethod
    def isomorphic(G1, G2):
        """Checks if two directed graphs are isomorphic."""
        return nx.is_isomorphic(G1, G2)

    @staticmethod
    def graph_power(G, k):
        """Returns the k-th power of a directed graph."""
        powered_graph = G.copy()
        for _ in range(k - 1):
            powered_graph = nx.compose(powered_graph, G)
        return powered_graph


# Testing the implemented algebraic operations
def test_graph_algebra():
    G1 = nx.DiGraph()
    G1.add_edges_from([("A", "B"), ("B", "C")])
    GraphView.plot_graph(G1, "G1")

    G2 = nx.DiGraph()
    G2.add_edges_from([("B", "C"), ("C", "D")])
    GraphView.plot_graph(G2, "G2")

    # Testing Union
    union_graph = GraphAlgebra.union(G1, G2)
    GraphView.plot_graph(union_graph, "Union of G1 and G2")

    # Testing Intersection
    intersection_graph = GraphAlgebra.intersection(G1, G2)
    GraphView.plot_graph(intersection_graph, "Intersection of G1 and G2")

    # Testing Difference
    difference_graph = GraphAlgebra.difference(G1, G2)
    GraphView.plot_graph(difference_graph, "Difference of G1 and G2")

    # Testing Complement
    complement_graph = GraphAlgebra.complement(G1)
    GraphView.plot_graph(complement_graph, "Complement of G1")

    # Testing Cartesian Product
    cartesian_graph = GraphAlgebra.cartesian_product(G1, G2)
    GraphView.plot_graph(cartesian_graph, "Cartesian Product of G1 and G2")

    # Testing Tensor Product
    tensor_graph = GraphAlgebra.tensor_product(G1, G2)
    GraphView.plot_graph(tensor_graph, "Tensor Product of G1 and G2")

    # Testing Isomorphic
    is_isomorphic = GraphAlgebra.isomorphic(G1, G2)
    print(f"G1 and G2 are isomorphic: {is_isomorphic}")

    # Testing Graph Power
    power_graph = GraphAlgebra.graph_power(G1, 2)
    GraphView.plot_graph(power_graph, "2nd Power of G1")

test_graph_algebra()
```

## Results from the algebraic operations on directed graphs:

1. **Union**: Combines both graphs \( G1 \) and \( G2 \), merging their nodes and edges.
2. **Intersection**: Retains only the nodes and edges common to both \( G1 \) and \( G2 \).
3. **Difference**: Removes the edges of \( G2 \) from \( G1 \) and also any isolated nodes resulting from this operation.
4. **Complement**: For the graph \( G1 \), it generates a graph containing all possible edges between its nodes except those already present in \( G1 \).
5. **Cartesian Product**: Combines nodes of \( G1 \) and \( G2 \) to form pairs and connects them based on the cartesian product definition.
6. **Tensor Product**: Combines nodes of \( G1 \) and \( G2 \) to form pairs and connects them based on the tensor product definition.
7. **Isomorphic**: This checks if \( G1 \) and \( G2 \) are structurally identical, considering node names. For the given example, they are not isomorphic.
8. **Graph Power**: It raises the graph \( G1 \) to the 2nd power, effectively connecting nodes that are at most 2 edges apart in the original graph.
