##Problem:
Let's define a "sevenish" number to be one which is either a power of 7, or the sum of unique powers of 7. The first few sevenish numbers are 1, 7, 8, 49, and so on. Create an algorithm to find the nth sevenish number.

##Solution:
To create an algorithm for finding the \( n \)th "sevenish" number, we can follow these steps:

1. **Initialize Variables**: Create a list to store the sevenish numbers and a counter to keep track of the number of sevenish numbers found.
2. **Generate Sevenish Numbers**: We can generate sevenish numbers by considering powers of 7. Each new power of 7 is a sevenish number. Additionally, we can create other sevenish numbers by adding unique combinations of previous powers of 7.
3. **Find the \( n \)th Sevenish Number**: Continue generating sevenish numbers until the \( n \)th number is found.

Here's how we can implement these steps:

1. Start with 1, which is \( 7^0 \), the first sevenish number.
2. For each new power of 7, add it to the list.
3. To generate other sevenish numbers, iterate through the list and add the current power of 7 to each existing number, ensuring each combination is unique.
4. Repeat this process until the list contains \( n \) numbers.
5. The \( n \)th number in the list is the \( n \)th sevenish number.

Let's implement this algorithm in Python.



```python
def nth_sevenish_number(n):
    # List to store sevenish numbers
    sevenish_numbers = []

    # Current power of 7
    power = 0

    # Generate sevenish numbers until we reach the nth number
    while len(sevenish_numbers) < n:
        # Next power of 7
        next_power_of_7 = 7 ** power

        # Generate new sevenish numbers by adding the current power of 7 to each existing number
        new_combinations = [next_power_of_7]
        for num in sevenish_numbers:
            new_combination = num + next_power_of_7
            new_combinations.append(new_combination)

        # Add the new combinations to the list of sevenish numbers
        sevenish_numbers.extend(new_combinations)

        # Increment the power for next iteration
        power += 1

    return sevenish_numbers[n - 1]

# Test the function with a few examples
nth_sevenish_number(1), nth_sevenish_number(2), nth_sevenish_number(3), nth_sevenish_number(4), nth_sevenish_number(5)
```


The corrected algorithm now produces the expected sequence of "sevenish" numbers. Here are the first five sevenish numbers:

1. For \( n = 1 \): The first sevenish number is 1.
2. For \( n = 2 \): The second sevenish number is 7.
3. For \( n = 3 \): The third sevenish number is 8 (which is 1 + 7).
4. For \( n = 4 \): The fourth sevenish number is 49 (which is \( 7^2 \)).
5. For \( n = 5 \): The fifth sevenish number is 50 (which is 1 + 49).

This algorithm effectively generates the \( n \)th sevenish number by considering both powers of 7 and unique sums of these powers.

##Generalising to Ishish numbers:
To generalize this algorithm for "ishish" numbers, where "ishish" is defined similarly to "sevenish" but with a different base number \( k \) instead of 7, we can modify the algorithm as follows:

1. **Parameterize the Base Number**: Instead of hardcoding the base as 7, we'll use a variable \( k \) to represent the base number.
2. **Generate Ishish Numbers**: Follow the same logic as with sevenish numbers, but use powers of \( k \) and their unique combinations.

Here's how the generalized algorithm would look in Python:



```python
def nth_ishish_number(k, n):
    # List to store ishish numbers
    ishish_numbers = []

    # Current power of k
    power = 0

    # Generate ishish numbers until we reach the nth number
    while len(ishish_numbers) < n:
        # Next power of k
        next_power_of_k = k ** power

        # Generate new ishish numbers by adding the current power of k to each existing number
        new_combinations = [next_power_of_k]
        for num in ishish_numbers:
            new_combination = num + next_power_of_k
            new_combinations.append(new_combination)

        # Add the new combinations to the list of ishish numbers
        ishish_numbers.extend(new_combinations)

        # Increment the power for next iteration
        power += 1

    return ishish_numbers[n - 1]

# Test the function with a few examples
nth_ishish_number(7, 1), nth_ishish_number(7, 2), nth_ishish_number(7, 3), nth_ishish_number(7, 4), nth_ishish_number(7, 5), nth_ishish_number(3, 4)
```

The generalized algorithm for finding the \( n \)th "ishish" number, where the base is specified by \( k \), seems to be working correctly. Here are some examples:

1. For base \( k = 7 \):
   - 1st ishish number: 1
   - 2nd ishish number: 7
   - 3rd ishish number: 8 (1 + 7)
   - 4th ishish number: 49 (\( 7^2 \))
   - 5th ishish number: 50 (1 + 49)

2. For base \( k = 3 \) and \( n = 4 \):
   - The 4th ishish number is 9 (\( 3^2 \)).

This approach allows you to specify any base number \( k \) and find the corresponding \( n \)th "ishish" number, based on powers of \( k \) and unique sums of these powers.

##Shpirals of Ishish numbers:
The concept of forming spirals with numbers is fascinating and is most famously exemplified by the Ulam spiral for prime numbers. To create a similar spiral with ishish numbers, we can plot these numbers on a two-dimensional grid, following a spiral pattern.

Here's how we can approach it:

1. **Generate a Large Set of Ishish Numbers**: We need a substantial list of ishish numbers to create a visually interesting spiral. We'll generate these numbers using the previously discussed algorithm.

2. **Plot the Numbers in a Spiral Pattern**: We'll start at the center of a grid and move in a spiral pattern (right, up, left, down, and so forth, increasing the step size appropriately). At each step, we'll mark the grid if the current number is an ishish number.

3. **Visual Representation**: We can use a scatter plot or a similar type of graph to visually represent the locations of ishish numbers on the grid.

For demonstration, let's create a spiral using a relatively small base, say 3, and plot the first few ishish numbers. We'll use Python and a plotting library like matplotlib for visualization.

Here is a spiral visualization of the first 200 "ishish" numbers for the base 3. In this spiral, each point represents an ishish number. The pattern starts from the center and spirals outward.

As you can see, unlike the Ulam spiral for primes, the distribution and pattern of ishish numbers appear to be different and less regular. This difference stems from the fundamental properties of ishish numbers, which are based on powers and sums of a base number, in contrast to primes which are defined by their indivisibility.

This visual exploration offers an interesting way to observe the properties and distributions of ishish numbers in a graphical format.

```python
import matplotlib.pyplot as plt

def generate_ishish_numbers(k, count):
    """ Generate a list of the first 'count' ishish numbers for base 'k'. """
    ishish_numbers = []
    power = 0

    while len(ishish_numbers) < count:
        next_power_of_k = k ** power
        new_combinations = [next_power_of_k]
        for num in ishish_numbers:
            new_combination = num + next_power_of_k
            new_combinations.append(new_combination)

        ishish_numbers.extend(new_combinations)
        power += 1

    return set(ishish_numbers[:count])

def plot_ishish_spiral(k, count):
    """ Plot a spiral of the first 'count' ishish numbers for base 'k'. """
    ishish_numbers = generate_ishish_numbers(k, count)

    # Initial position
    x, y = 0, 0
    dx, dy = 1, 0
    grid_size = int(count ** 0.5) + 1

    x_values, y_values = [], []

    for i in range(1, count + 1):
        if i in ishish_numbers:
            x_values.append(x)
            y_values.append(y)

        # Move to next position
        if (abs(x) == abs(y) and x != y) or (x > 0 and x == 1 - y) or (x < 0 and x == -y):
            dx, dy = -dy, dx  # Change direction

        x, y = x + dx, y + dy

    plt.figure(figsize=(10, 10))
    plt.scatter(x_values, y_values, color='blue')
    plt.title(f'Ishish Spiral for Base {k}')
    plt.axis('equal')
    plt.axis('off')
    plt.show()

# Plot the spiral for base 3 with the first 200 ishish numbers
plot_ishish_spiral(3, 200)
```

##Spirally Shpiral Ishish numbers:
To create more traditional-looking spirals ("shpirals") using ishish numbers, we can map each ishish number to polar coordinates (radius \( r \) and angle \( \theta \)). In a typical spiral, the radius increases as the angle increases. For ishish shpirals, we can use the sequence of ishish numbers to dictate the increase in radius, while \( \theta \) can be a function of the position of each number in the sequence.

1. **Radius (\( r \))**: Let the radius increase with each ishish number. The value of each ishish number can determine the radius, possibly scaled down to keep the spiral manageable in size.

2. **Angle (\( \theta \))**: The angle can increase linearly or exponentially with each step. A linear increase will result in a more evenly spaced spiral, while an exponential increase will create more tightly wound spirals at the beginning.

3. **Plotting**: Convert these polar coordinates to Cartesian coordinates for plotting.

Let's implement this approach to generate and plot a shpiral for a given base \( k \). We'll use a linear increase in \( \theta \) for simplicity. I'll demonstrate this with a base of 3 as an example.



```python
import numpy as np

def generate_ishish_shpiral(k, count):
    """ Generate and plot a shpiral for ishish numbers with base 'k'. """
    ishish_numbers = list(generate_ishish_numbers(k, count))

    # Constants for the spiral
    angle_increment = 0.1  # Increment of angle for each step
    radius_scaling = 0.1   # Scaling factor for radius

    # Convert ishish numbers to polar coordinates
    angles = np.arange(0, angle_increment * count, angle_increment)
    radii = [num * radius_scaling for num in ishish_numbers]

    # Convert polar coordinates to Cartesian coordinates for plotting
    x_values = [r * np.cos(theta) for r, theta in zip(radii, angles)]
    y_values = [r * np.sin(theta) for r, theta in zip(radii, angles)]

    return x_values, y_values

# Plot the shpiral for base 3
plt.figure(figsize=(8, 8))
x_values, y_values = generate_ishish_shpiral(3, 200)
plt.scatter(x_values, y_values, color='blue')
plt.title('Ishish Shpiral for Base 3')
plt.axis('equal')
plt.show()

```

Here is the "shpiral" for ishish numbers with base 3, plotted using polar coordinates. In this visualization:

- The radius \( r \) is determined by the value of each ishish number, scaled for better visibility.
- The angle \( \theta \) increases linearly with each step.

This approach creates a spiral pattern where the distribution of points reflects the unique properties of ishish numbers. The spacing and arrangement of the points in the shpiral can provide insights into the nature of these numbers and their progression.

Such visualizations can be an interesting addition to our paper, offering a more traditional spiral representation of ishish numbers while highlighting their distinct characteristics.

```python
# Adjusting the code to generate shpirals for multiple bases using the new approach

def generate_ishish_shpiral_plots(bases, count):
    angle_increment = 0.1  # Increment of angle for each step
    radius_scaling = 0.1   # Scaling factor for radius

    # Set up the plot
    plt.figure(figsize=(12, 12))

    for i, base in enumerate(bases, 1):
        ishish_numbers = list(generate_ishish_numbers(base, count))
        angles = np.arange(0, angle_increment * count, angle_increment)
        radii = [num * radius_scaling for num in ishish_numbers]

        # Convert polar coordinates to Cartesian coordinates for plotting
        x_values = [r * np.cos(theta) for r, theta in zip(radii, angles)]
        y_values = [r * np.sin(theta) for r, theta in zip(radii, angles)]

        # Plot each shpiral
        plt.subplot(2, 2, i)
        plt.scatter(x_values, y_values, color='blue')
        plt.title(f'Ishish Shpiral for Base {base}')
        plt.axis('equal')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Selected bases for shpirals
bases = [2, 3, 5, 8]

# Generating and plotting shpirals for the selected bases
generate_ishish_shpiral_plots(bases, 200)

```

Here are the shpirals for the bases 2, 3, 5, and 8, each created using the improved approach with polar coordinates. Each subplot represents a shpiral for a specific base, and the patterns of the shpirals vary with the base number.

These visualizations demonstrate the unique characteristics of ishish numbers in a more traditional spiral form. The way the points are distributed in each shpiral reflects the mathematical properties of ishish numbers for each base.

##Shpirals in 3d:
Extending the shpirals into a 3D representation by incorporating the Z-axis is an interesting idea. In a 3D shpiral, we can use the third dimension to further differentiate the ishish numbers. One approach is to let the Z-axis represent the sequence of the ishish numbers, so the shpiral grows upwards (or downwards) as it spirals out.

Here's how we can modify the approach:

1. **X and Y Coordinates**: Continue using the polar coordinates to determine the X and Y positions in the spiral pattern.

2. **Z Coordinate**: Increment the Z coordinate linearly or based on some function of the ishish number's position in the sequence. This will create a helical, 3D effect.

3. **3D Plotting**: We'll use a 3D plotting library, such as `matplotlib`'s 3D toolkit, to visualize this 3D shpiral.

Let's implement this to visualize a 3D shpiral for a chosen base number, say 3. We will plot the first 200 ishish numbers in this 3D space.



```python
from mpl_toolkits.mplot3d import Axes3D

def generate_3d_ishish_shpiral(k, count):
    ishish_numbers = list(generate_ishish_numbers(k, count))

    angle_increment = 0.1  # Increment of angle for each step
    radius_scaling = 0.1   # Scaling factor for radius
    z_scaling = 0.1        # Scaling factor for z-axis

    # Convert ishish numbers to polar coordinates
    angles = np.arange(0, angle_increment * count, angle_increment)
    radii = [num * radius_scaling for num in ishish_numbers]
    z_values = [i * z_scaling for i in range(count)]  # Linear increase along z-axis

    # Convert polar coordinates to Cartesian coordinates for plotting
    x_values = [r * np.cos(theta) for r, theta in zip(radii, angles)]
    y_values = [r * np.sin(theta) for r, theta in zip(radii, angles)]

    return x_values, y_values, z_values

# Plot the 3D shpiral for base 3
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

x_values, y_values, z_values = generate_3d_ishish_shpiral(3, 200)
ax.scatter(x_values, y_values, z_values, c=z_values, cmap='viridis')

ax.set_title('3D Ishish Shpiral for Base 3')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()

```

Here is the 3D representation of a shpiral for ishish numbers with base 3. This visualization takes the concept of the shpiral into three dimensions:

- The X and Y coordinates form the spiral pattern based on polar coordinates.
- The Z coordinate grows linearly, adding a helical dimension to the shpiral.

This 3D shpiral offers a more dynamic and intricate view of the ishish numbers, providing an even deeper insight into their unique distribution and properties. Such a representation could be a visually striking and informative addition to our paper, showcasing the versatility and richness of ishish number patterns.
