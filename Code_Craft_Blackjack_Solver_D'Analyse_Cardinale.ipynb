##Problem:
Blackjack is a two player card game whose rules are as follows:

The player and then the dealer are each given two cards.
The player can then "hit", or ask for arbitrarily many additional cards, so long as their total does not exceed 21.
The dealer must then hit if their total is 16 or lower, otherwise pass.
Finally, the two compare totals, and the one with the greatest sum not exceeding 21 is the winner.
For this problem, cards values are counted as follows: each card between 2 and 10 counts as their face value, face cards count as 10, and aces count as 1.

```python
def blackjack_solver(deck):
    def play_hand(deck, player_turn=True):
        score = 0
        while True:
            # Player or dealer stops hitting if score > 16 or deck is empty
            if score > 16 or not deck:
                break
            # Player's strategy: hit if next card doesn't bust them and it's their turn
            if player_turn and score + deck[0] <= 21:
                score += deck.pop(0)
            # Dealer's turn, hit if score <= 16
            elif not player_turn and score <= 16:
                score += deck.pop(0)
            else:
                break
        return score

    wins = 0
    losses = 0

    while len(deck) >= 4:  # Need at least 4 cards to play a hand
        player_score = play_hand(deck[:2] + deck[4:], player_turn=True)
        dealer_score = play_hand(deck[2:4] + deck[4:], player_turn=False)
        deck = deck[4:]  # Remove the top 4 cards used in this round

        # Scoring the game
        if player_score > 21:
            losses += 1  # Player busts
        elif dealer_score > 21 or player_score > dealer_score:
            wins += 1  # Dealer busts or player has higher score
        elif player_score < dealer_score:
            losses += 1  # Dealer has higher score

    return wins - losses

# Example deck
example_deck = [10, 6, 5, 10, 3, 2, 7, 8, 9, 1, 10, 4, 10, 10, 10, 3, 2, 1, 8, 7, 6, 5, 4, 3, 2, 10, 10, 10, 10, 1]
blackjack_solver(example_deck)
```

```python
def blackjack_basic_strategy_simulation():
    basic_strategy_chart = {
        # Format: (player_total, dealer_show_card): action
        # Actions: 'H' for Hit, 'S' for Stand
        # Hard totals
        (8, '2'): 'H', (8, '3'): 'H', (8, '4'): 'H', (8, '5'): 'H', (8, '6'): 'H', (8, '7'): 'H', (8, '8'): 'H', (8, '9'): 'H', (8, '10'): 'H', (8, 'A'): 'H',
        # ... continue for other hard totals
        # Soft totals (Ace counted as 11)
        (13, '2'): 'H', (13, '3'): 'H', (13, '4'): 'H', (13, '5'): 'H', (13, '6'): 'H', (13, '7'): 'H', (13, '8'): 'H', (13, '9'): 'H', (13, '10'): 'H', (13, 'A'): 'H',
        # ... continue for other soft totals
    }

    def card_value(card):
        """Returns the Blackjack value of a card."""
        return min(card, 10)  # Face cards are valued at 10

    def is_soft_hand(hand):
        """Check if the hand is a soft hand."""
        return 1 in hand and sum(map(card_value, hand)) + 10 <= 21

    def get_action(player_hand, dealer_show_card):
        """Determine the best action based on the basic strategy chart."""
        total = sum(map(card_value, player_hand))
        if is_soft_hand(player_hand):
            total += 10  # Count Ace as 11

        action = basic_strategy_chart.get((total, str(dealer_show_card)), 'S')  # Default to 'Stand'
        return action

    def play_hand():
        player_hand = [random.randint(1, 13), random.randint(1, 13)]
        dealer_show_card = random.randint(1, 13)

        print(f"Your hand: {player_hand}, Dealer's show card: {dealer_show_card}")
        action = get_action(player_hand, dealer_show_card)
        print(f"Basic Strategy suggests: {'Hit' if action == 'H' else 'Stand'}")

    return play_hand

# Run the basic strategy simulation
basic_strategy_simulation = blackjack_basic_strategy_simulation()
basic_strategy_simulation()  # Simulates a hand and suggests the best action based on the basic strategy

```

### Part 3: Card Counting Basics

**Objective**: Learn the basics of the Hi-Lo card counting system, a simple yet effective method.

#### Key Concepts:

1. **Hi-Lo System**: This system assigns a value of +1, 0, or -1 to each card seen. Low cards (2-6) are +1, neutral cards (7-9) are 0, and high cards (10-Ace) are -1.

2. **Running Count**: You keep a "running count" based on the cards dealt. This count reflects the ratio of high to low cards.

3. **True Count**: In multi-deck games, you convert the running count into a "true count" by dividing the running count by the number of decks remaining.

4. **Betting Strategy**: You adjust your bets based on the true count. Higher counts favor the player, so you bet more.

5. **Playing Decisions**: The count can also influence playing decisions, especially in borderline cases.
