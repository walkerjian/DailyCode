##Problem:
A number is considered perfect if its digits sum up to exactly 10.

Given a positive integer n, return the n-th perfect number.

For example, given 1, you should return 19. Given 2, you should return 28.

##Solution:
To find the n-th perfect number where the sum of its digits equals 10, we can start with the smallest perfect number, which is 19, and then incrementally find the next perfect numbers. Here is how we can do it:

1. Start with the smallest perfect number, 19.
2. Increment the number by 1 each time.
3. Check if the sum of the digits equals 10. If so, it is a perfect number.
4. Continue until we find the n-th perfect number.




##Implementation:
Let's implement this algorithm to find the n-th perfect number.

The n-th perfect number function works as expected. For $ n = 1 $, it returns 19, and for $ n = 2 $, it returns 28. Thus, you can use this function to find any n-th perfect number where the sum of its digits equals 10.


```python
def find_nth_perfect_number(n):
    count = 0  # Counter for how many perfect numbers have been found
    num = 18   # Start from 18, as 19 is the first perfect number

    while count < n:
        num += 1  # Increment the number to check
        # Sum the digits of the current number
        if sum(map(int, str(num))) == 10:
            count += 1  # If the number is perfect, increment the count

    return num

# Testing the function with the provided examples
nth_1 = find_nth_perfect_number(1)
nth_2 = find_nth_perfect_number(2)
nth_1, nth_2
```

```python
import matplotlib.pyplot as plt

# Generate a list of the first 100 perfect numbers
perfect_numbers = [find_nth_perfect_number(i + 1) for i in range(100)]

# Plot the perfect numbers
plt.figure(figsize=(10, 6))
plt.plot(perfect_numbers, marker='o', linestyle='-')
plt.title('Distribution of the First 100 Perfect Numbers')
plt.xlabel('Ordinal Position (n-th Perfect Number)')
plt.ylabel('Value of Perfect Number')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to find perfect numbers in a given base
def find_perfect_numbers_in_base(base, count):
    perfect_numbers = []
    num = 1  # Start with 1 as the smallest number to check
    while len(perfect_numbers) < count:
        # Convert the number to the given base and sum its digits
        digit_sum = sum(int(digit, base) for digit in np.base_repr(num, base=base))
        if digit_sum == base:
            perfect_numbers.append(num)
        num += 1
    return perfect_numbers

# We will find the first few perfect numbers in each base
count = 20  # Number of perfect numbers to find in each base

# Generating perfect numbers for bases 2 to 16
perfect_numbers_bases = {base: find_perfect_numbers_in_base(base, count) for base in range(2, 17)}

# Plotting in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

for base, numbers in perfect_numbers_bases.items():
    ax.scatter([base] * len(numbers), list(range(len(numbers))), numbers, label=f'Base {base}')

ax.set_xlabel('Base')
ax.set_ylabel('Ordinal Position')
ax.set_zlabel('Perfect Number Value')
ax.set_title('Perfect Numbers in Different Bases')
ax.legend()

plt.show()
```
