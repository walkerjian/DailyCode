It is heap-ordered, so that each parent value is strictly less than that of its children.
An in-order traversal of the tree produces nodes with values that correspond exactly to S.
For example, given the sequence [3, 2, 6, 1, 9], the resulting Cartesian tree would be:
````
      1
    /   \   
  2       9
 / \
3   6
````

    def _build_tree(self, sequence):
        """
        Recursively build the Cartesian tree.
        
        The idea:
        1. Find the minimum value in the sequence.
        2. Create a new node with this value.
        3. Recursively build the left subtree with the sequence before this minimum value.
        4. Recursively build the right subtree with the sequence after this minimum value.
        5. Return the current node.

        return root
