For example, given A = 3 -> 7 -> 8 -> 10 and B = 99 -> 1 -> 8 -> 10, return the node with value 8.

In this example, assume nodes with the same value are the exact same node objects.

### Model:

Now, let's dive into the logic of finding the intersection point of two linked lists.

The basic idea is:

1. Traverse both linked lists to determine their lengths.
2. Calculate the difference in lengths of the two linked lists (let's say `d`).
3. Move the head of the longer linked list `d` steps forward.
4. Now, traverse both linked lists simultaneously until we find a common node or reach the end.

```python
def get_intersection_point(list1, list2):
    """Find the intersection point of two singly linked lists."""
    
    def get_length(node):
        """Get the length of the linked list starting from the node."""
        count = 0
        while node:
            count += 1
            node = node.next
        return count
    
    len1, len2 = get_length(list1), get_length(list2)
    diff = abs(len1 - len2)

    # Move the head of the longer list 'diff' steps forward
    if len1 > len2:
        for _ in range(diff):
            list1 = list1.next
    else:
        for _ in range(diff):
            list2 = list2.next

    # Traverse both lists simultaneously
    while list1 and list2:
        if list1 == list2:
            return list1.value
        list1, list2 = list1.next, list2.next

    return None
```
